1. What is Python, and why is it popular?
   - Python is a high-level, interpreted programming language that is known for its simple and easy-to-read syntax. It is widely used in different fields such as web development, data science, machine learning, automation, scientific computing, and software development.

   The main reasons for Python’s popularity are:
   
   1. Easy to learn and use :- Python has a clean and simple syntax that looks like English, which makes it beginner-friendly.
   2. Large community support :- It has a huge community, so we can easily find tutorials, documentation, and solutions to problems.
   3. Rich libraries and frameworks :- Python provides powerful libraries like NumPy, Pandas, Matplotlib, Scikit-learn, and TensorFlow, which make tasks easier, especially in data science and machine learning.
   4. Platform independent :- Python code can run on Windows, macOS, and Linux without major changes.
   5. Faster development :- Python requires fewer lines of code compared to other languages, which helps in rapid development.

2. What is an interpreter in Python?
   - A Python interpreter is a program that runs Python code by reading it line by line and giving the output.It takes the code we write, checks if there are any errors, and if everything is correct, it executes the code step by step. Because of this, we don't need to compile the whole program before running it.
   This makes Python easy to use and beginner-friendly because
     - We can run the code instantly
     - Errors are shown clearly, so they are easy to fix
     -  We can execute one line at a time in interactive mode.

3. What are pre-defined keywords in Python?
   - Pre-defined keywords in Python are special reserved words that already
   have a fixed meaning in the Python language.We cannot use these words as variable names, function names, or identifiers because Python uses them for its own syntax and structure.These keywords help us write logical statements, control the flow of the program, and define functions, classes, loops, and conditions.
     - Example- if, else, elif, for, while, break, continue, def, return, class, try, except, import, from, pass, True, False, None, etc.


4. Can keywords be used as variable names
   - No, keywords cannot be used as variable names in Python because they are reserved words with a special meaning in the language.

5. What is mutability in Python?
   - Mutability in Python means whether an object's value can be changed after it is created. If the value of an object can be modified, it is called mutable. In these objects, we can change, add, or remove elements without creating a new object. Examples: list, dictionary, set
   

```python
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)
```

6. Why are lists mutable, but tuples are immutable?
   - Lists are mutable and tuples are immutable because they are designed for different purposes in Python. A list is used when we need a collection of items that can be changed. We can add, remove, or update elements in a list after it is created.

   - A tuple is used when we want a collection of items that should not be changed.Once a tuple is created, its elements cannot be added, removed, or updated.
   

```python
# list Example
my_list = [1, 2, 3]
my_list[1] = 10
print(my_list)
# the original list is modified.

#Tuple Example
my_tuple = (1, 2, 3)
my_tuple[1] = 10
#This will give an error because tuples are immutable
```

```python

# 7. What is the difference between a list and a tuple in Python?

"""
The == and is operators in Python are both used for comparison, but they
check different things.

== (Equality operator)

* It checks whether the values of two objects are equal or not.
* It compares the data inside the objects.
"""
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)

"""
is (Identity operator)

* It checks whether both variables point to the same object in memory.
* It compares the memory location, not the value.

"""
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)

"Example where is becomes True"

a = [1, 2, 3]
b = a

print(a is b)

```

```python
# 8. What are logical operators in Python?
"""
Logical operators in Python are used to perform logical operations on boolean
values.
The three logical operators in Python are:
    1. and
    2. or
    3. not

1. and operator:- The and operator returns True if both operands are True,
otherwise it returns False.
"""
x = 10
print(x > 5 and x < 20)

"""
2. or operator:- The or operator returns True if at least one of the operands
is True, otherwise it returns False.
"""
x = 5
print(x > 3 or x < 1)

"""
3. not operator:- The not operator returns the opposite of the operand. If the
operand is True, it returns False, and if the operand is False, it returns True.
"""
x = 5
print(not(x > 3))

"""
Logical operators are used when we want to check multiple conditions together:

and - both conditions must be true

or - at least one condition must be true

not - gives the opposite result
"""
```

```python
# 9. What is type casting in Python?

"""
Type casting in Python means converting one data type into another data type.
Sometimes we need to change the type of a variable so that we can perform a
specific operation. This conversion is called type casting.

Common type casting functions:

1. int() → converts a value to integer
2. float() → converts a value to float
3. str() → converts a value to string
4. list(), tuple(), set() → converts one collection type into another
"""
x = "10"
y = int(x)

print(y)
print(type(y))
```

```python
# 10. What is the difference between implicit and explicit type casting?

"""
The difference between implicit and explicit type casting is based on who
 performs the conversion — Python automatically or the programmer manually.

1. Implicit Type Casting

In implicit type casting, Python automatically converts one data type into
another without the user’s involvement.
It usually happens when there is no risk of data loss.
"""
a = 5      # int
b = 2.5    # float

c = a + b
print(c)
print(type(c))

"""
Explicit Type Casting

In explicit type casting, the programmer manually converts the data type using
functions like int(), float(), str(), etc.
"""
x = "10"
y = int(x)

print(y)
print(type(y))
```

```python
# 11. What is the purpose of conditional statements in Python?

"""
The purpose of conditional statements in Python is to make decisions in a
program based on certain conditions.They allow the program to execute
different blocks of code depending on whether a condition is True or False.
This helps in controlling the flow of the program.

For example

* check if a number is positive or negative

* give output based on a user’s input

* run a particular block of code only when a condition is satisfied
"""
age = 18

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")

"""Here, the program checks the condition.
If it is True, the first block runs; otherwise, the second block runs."""
```

```python
# 12. How does the elif statement work?

"""
The elif statement in Python stands for “else if”.
It is used to check multiple conditions one by one after an if statement.

When Python finds a condition that is True, it executes that block and skips
the remaining conditions.

if condition1:
    # code
elif condition2:
    # code
elif condition3:
    # code
else:
    # code

Example:-
"""
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# 13. What is the difference between for and while loops

"""
The for loop and while loop are both used for repetition in Python, but they
are used in different situations.

For loop:-A for loop is used when we already know how many times we want to
run the loop or when we are iterating over a sequence like a list, tuple,
string, or range.
"""
for i in range(5):
    print(i)
"This loop will run 5 times because the range is already defined."

"""
While loop:-A while loop is used when we don’t know the exact number of
iterations in advance.
It runs as long as the given condition is True.
"""
i = 0
while i < 5:
    print(i)
    i += 1

"This loop will continue until the condition becomes False."
```

```python
# 14. Describe a scenario where a while loop is more suitable than a for loop.

"""
A while loop is more suitable when we do not know in advance how many times
the loop should run and the execution depends on a condition.

Scenario:
Suppose we are taking input from a user and the loop should continue until
the user enters a correct password.

In this case, we don’t know how many attempts the user will take,
so a while loop is the better choice.
"""
password = ""

while password != "python123":
    password = input("Enter the password: ")

print("Access granted")
```

```python
# 14. Write a program to reverse a string using a while loop.
text = input("Enter a string: ")

reversed_text = ""
i = len(text) - 1

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))

fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial of", num, "is:", fact)
```
